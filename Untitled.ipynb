```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler  = StandardScaler()
X = scaler.fit_transform(x)
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC, NuSVC
from sklearn.tree import DecisionTreeClassifier 
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import  confusion_matrix
from sklearn.metrics import plot_confusion_matrix
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
lr = LogisticRegression(max_iter = 500)
clf = lr.fit(X_train, y_train)
```

```python
predictions_lr = lr.predict(X_test)
for i in predictions_lr:
    print(i)
```

```python
score = lr.score(X_test, y_test)
print(score*100)
```

```python
plot_confusion_matrix(lr, X_test, y_test)
```

```python
for i in range (1,15,2):
 knn2 = KNeighborsClassifier(n_neighbors=i)
 knn2.fit(X_train, y_train)
 print("For k = %d accuracy is"%i,knn2.score(X_test,y_test))
```

```python
plot_confusion_matrix(knn2, X_test, y_test)
```

```python
dtree = DecisionTreeClassifier()
dtree = dtree.fit(X_train, y_train)
```

```python
print("Decision Tree Test Accuracy {:.2f}%".format(dtree.score(X_test,y_test)*100))
```

```python
plot_confusion_matrix(dtree,X_test, y_test)
```

```python
rf = RandomForestClassifier(n_estimators = 1000, random_state = 1)
rf.fit(X_train, y_train)
print("Random Forest Algorithm Accuracy Score : {:.2f}%".format(rf.score(X_test,y_test)*100))
```

```python
plot_confusion_matrix(rf,X_test, y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
nb = GaussianNB()
nb.fit(X_train, y_train)
print("Accuracy of Naive Bayes: {:.2f}%".format(nb.score(X_test,y_test)*100))
```

```python
plot_confusion_matrix(nb, X_test, y_test)
```

```python
gcf = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,
                                 max_depth=1, random_state=0).fit(X_train, y_train)
clf.score(X_test,y_test)
print("Accuracy of gradient boosting: {:.2f}%".format(gcf.score(X_test,y_test)*100))
```

```python
plot_confusion_matrix(gcf, X_test, y_test)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
ext = ExtraTreesClassifier(n_estimators=10, max_depth=None,min_samples_split=2, random_state=0)
ext.fit(X_train, y_train)
print("Extra Trees Algorithm Accuracy Score : {:.2f}%".format(ext.score(X_test,y_test)*100))
```

```python
plot_confusion_matrix(ext, X_test, y_test)
```
